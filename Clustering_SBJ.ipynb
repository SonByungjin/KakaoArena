```python
import pandas as pd
df = pd.read_json('c:/data/kakaoarena/train.json',encoding='utf-8')
df
```

```python
docs=[]
for i in df['tags']:
    docs.append(i)
```

```python
docs
```

```python
for i in range(len(docs)):
    docs[i]=' '.join(docs[i])
```

```python
docs = docs[0:1000]
```

```python
docs
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec=CountVectorizer()
X=vec.fit_transform(docs) 
df2=pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
df2.head()
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5, random_state=10).fit(df2)
kmeans.labels_
```

```python
%matplotlib inline
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
#주성분분석 n_components 성분의 갯수
#플로팅을 위하여 3개의 변수를 2개의 주성분으로 축소(차원축소)
pca=PCA(n_components=2, random_state=10)
components=pca.fit_transform(df2)
df3=pd.DataFrame(data=components, columns=['comp 1','comp 2'])
print(df3)
#kmeans.labels_==0
plt.scatter(df3.iloc[kmeans.labels_ == 0,0],
df3.iloc[kmeans.labels_ == 0,1],s=10,c='red',label='cluster 1')

plt.scatter(df3.iloc[kmeans.labels_ == 1,0],
df3.iloc[kmeans.labels_ == 1,1],s=10,c='blue',label='cluster 2')

plt.scatter(df3.iloc[kmeans.labels_ == 2,0],
df3.iloc[kmeans.labels_ == 2,1],s=10,c='green',label='cluster 3')

plt.scatter(df3.iloc[kmeans.labels_ == 3,0],
df3.iloc[kmeans.labels_ == 2,1],s=10,c='yellow',label='cluster 4')

plt.legend()
```
