```python
import pandas as pd
df = pd.read_json('c:/data/kakaoarena/train.json',encoding='utf-8')
df
```

```python
df=pd.DataFrame(df)
```

```python
tags_dataset = df['tags'].values
```

- 데이터프레임에서 칼럼을 지정하고 value를 붙여야 이중리스트 형태가 된다고 하시더라구요.
- 그리고 각 플레이리스트별 태그 묶음으로 인코딩을 진행했습니다.

```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

te = TransactionEncoder()
te_ary = te.fit(tags_dataset).transform(tags_dataset)
te_ary
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

- EDA에서 봤는데 플레이리스트별 태그 중에 40%정도만 중복된 태그라고 하더라구요
- 그래서 거의 다 False로 나오는 것 같습니다
- 텍스트 마이닝으로 비슷한 단어를 매핑할 수 있지 않을까 조심스럽게 생각해봅니다

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets
```

- 데이터에 비해 중복되는 경우의 수가 적다보니 빈도수(support)를 0.01까지 낮춰야 결과값이 나타났습니다.

```python
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_itemsets, metric="lift", min_threshold=3)
rules
```

- 태그 간의 상관관계 수치가 중요하다고 생각하여 lift(min_threshold) 수치를 조절하면서 결과를 냈습니다.
- 만약 연관규칙분석을 활용한다면 어떤 플레이리스트와 상관관계 높은 2-3가지 태그를 활용하여 다른 플레이리스트나 곡을 가져오게끔 할것 같습니다.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
```
